```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure visualizations render correctly in Jupyter
%matplotlib inline

# Set visualization style
sns.set(style="whitegrid")
```

```python
# Load the campaign performance dataset
campaign_df = pd.read_csv('cleaned_campaign_performance.csv')

# Load the creator profile dataset
creator_df = pd.read_csv('cleaned_dataset_creator_profile.csv')
```

```python
plt.figure(figsize=(10, 6))
platform_engagement = campaign_df.groupby('Platform')['Engagement Rate'].mean().sort_values(ascending=False)
sns.barplot(x=platform_engagement.index, y=platform_engagement.values, palette="Blues_d")
plt.title('Average Engagement Rate by Platform')
plt.ylabel('Engagement Rate (%)')
plt.xlabel('Platform')
plt.xticks(rotation=45)
plt.show()
```

```python
top_accounts = campaign_df.sort_values('Total Engagements', ascending=False).head(5)
plt.figure(figsize=(10, 6))
sns.barplot(x='Account', y='Total Engagements', data=top_accounts, palette="Greens_d")
plt.title('Top 5 Accounts by Total Engagements')
plt.ylabel('Total Engagements')
plt.xlabel('Account')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(creator_df['CREDIBILITY SCORE (fraud detection)'], bins=20, kde=True, color='coral')
plt.title('Distribution of Credibility Scores')
plt.xlabel('Credibility Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Convert percentage strings to float for calculation
gender_df = creator_df[['males', 'females']].replace('%','', regex=True).astype(float)

# Sum and plot
gender_df.sum().plot(kind='pie', autopct='%1.1f%%', startangle=140, figsize=(8, 8), colors=['lightblue', 'lightgreen'])
plt.title('Overall Audience Gender Distribution')
plt.ylabel('')
plt.show()
```

```python
top_countries = creator_df['top_country'].value_counts().head(5)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_countries.index, y=top_countries.values, palette="Oranges_d")
plt.title('Top 5 Countries by Creator Audience')
plt.ylabel('Number of Creators')
plt.xlabel('Country')
plt.show()
```
